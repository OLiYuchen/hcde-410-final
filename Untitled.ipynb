```python
# Calculate the conditional probability of having a factor (e.g., pet allergy) given asthma diagnosis
# Group the data by 'Diagnosis' (Asthma status) and calculate proportions for Pet Allergy

# Probability of having pet allergy given asthma
pet_allergy_given_asthma = data[data['Diagnosis'] == 1]['PetAllergy'].value_counts(normalize=True)

# Probability of having pet allergy given no asthma
pet_allergy_given_no_asthma = data[data['Diagnosis'] == 0]['PetAllergy'].value_counts(normalize=True)

# Combine the probabilities into a single DataFrame for comparison
probabilities = pd.DataFrame({
    'Pet Allergy with Asthma': pet_allergy_given_asthma,
    'Pet Allergy without Asthma': pet_allergy_given_no_asthma
}).fillna(0)  # Fill missing values with 0

# Sort by probability of having pet allergy with asthma in descending order
probabilities = probabilities.sort_values(by='Pet Allergy with Asthma', ascending=False)

# Display the probabilities
probabilities

```

```python
import matplotlib.pyplot as plt

# Create a bar chart for visualization
probabilities.plot(kind='bar', figsize=(10, 6))
plt.title('Conditional Probabilities: Pet Allergy and Asthma')
plt.ylabel('Probability')
plt.xlabel('Pet Allergy Status (0 = No, 1 = Yes)')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Required Imports
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace with your dataset file path if needed)
file_path = 'asthma_disease_data.csv'  # Adjust file path
data = pd.read_csv(file_path)

# Drop irrelevant columns and prepare data
data_cleaned = data.drop(columns=['DoctorInCharge', 'PatientID'])
X = data_cleaned.drop(columns=['Diagnosis'])
y = data_cleaned['Diagnosis']

def plot_confusion_matrix_for_feature(feature, X, y):
    # Prepare data for the single feature
    X_feature = X[[feature]]  # Select only the specific feature

    # Split into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X_feature, y, test_size=0.3, random_state=42)

    # Train Logistic Regression on the single feature
    model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
    model.fit(X_train, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test)

    # Generate and plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Asthma", "Asthma"])
    disp.plot(cmap='Blues', values_format='d')
    plt.title(f'Confusion Matrix for {feature}')
    plt.show()

# Features to analyze
features_to_analyze = ['PetAllergy', 'Smoking', 'FamilyHistoryAsthma', 'HistoryOfAllergies', 'Eczema', 'HayFever', 'GastroesophagealReflux']

# Loop through features and generate confusion matrices
for feature in features_to_analyze:
    print(f"Confusion Matrix for {feature}:")
    plot_confusion_matrix_for_feature(feature, data_cleaned, y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations

# List of categorical factors in the dataset
categorical_factors = ['PetAllergy', 'Smoking', 'FamilyHistoryAsthma', 'HistoryOfAllergies', 'Eczema', 'HayFever', 'GastroesophagealReflux']

# Create a dictionary to store probabilities
probabilities = {}

# Calculate probabilities for individual factors
for factor in categorical_factors:
    filtered_data = data_cleaned[data_cleaned[factor] == 1]
    probabilities[factor] = filtered_data['Diagnosis'].mean()

# Calculate probabilities for all pairs of factors
for factor1, factor2 in combinations(categorical_factors, 2):
    filtered_data = data_cleaned[(data_cleaned[factor1] == 1) & (data_cleaned[factor2] == 1)]
    probabilities[f"{factor1} & {factor2}"] = filtered_data['Diagnosis'].mean()

# Convert the probabilities dictionary into a DataFrame
probabilities_df = pd.DataFrame.from_dict(probabilities, orient='index', columns=['Probability'])
probabilities_df.reset_index(inplace=True)
probabilities_df.rename(columns={'index': 'Factor'}, inplace=True)

# Convert probabilities to percentages
probabilities_df['Probability'] = probabilities_df['Probability'] * 100

# Sort probabilities for better visualization
probabilities_df.sort_values(by='Probability', ascending=False, inplace=True)

# Plot a bar chart
plt.figure(figsize=(12, 8))
bars = plt.bar(probabilities_df['Factor'], probabilities_df['Probability'], color='skyblue', edgecolor='black')

# Add value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, yval + 0.5, f'{yval:.1f}%', ha='center', va='bottom')

plt.title('Probability of Asthma Diagnosis for Categorical Factors and Their Combinations')
plt.ylabel('Probability (%)')
plt.xlabel('Factors')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
