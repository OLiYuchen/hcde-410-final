## Part 2: Investigating relationship of different factors with of Asthma Diagnosis
### Part 2.a In this section, we will calculate the probabilities of asthma diagnosis for:
1. Individual categorical factors (e.g., PetAllergy, Smoking, etc.).
2. Combinations of two categorical factors (e.g., PetAllergy & Smoking).
The results will be visualized in a bar chart for better interpretation.

#### Define Factors and Initialize Storage

```python
# List of categorical factors in the dataset
categorical_factors = ['PetAllergy', 'Smoking', 'FamilyHistoryAsthma', 
                       'HistoryOfAllergies', 'Eczema', 'HayFever', 'GastroesophagealReflux']

# Dictionary to store calculated probabilities
probabilities = {}
```

#### Calculating Probabilities for Individual Categorical Factors
For each categorical factor, we filter the data where the factor is `1` (true) 
and calculate the mean asthma diagnosis rate.

#### Calculating Probabilities for Factor Combinations
For each pair of categorical factors, we filter the data where both factors 
are `1` (true) and calculate the mean asthma diagnosis rate.

```python
# Calculate probabilities for individual factors
for factor in categorical_factors:
    filtered_data = data[data[factor] == 1]
    probabilities[factor] = filtered_data['Diagnosis'].mean()

# Calculate probabilities for all pairs of factors
for factor1, factor2 in combinations(categorical_factors, 2):
    filtered_data = data[(data[factor1] == 1) & (data[factor2] == 1)]
    probabilities[f"{factor1} & {factor2}"] = filtered_data['Diagnosis'].mean()

```

#### Visualizing Probabilities
The calculated probabilities will be visualized in a bar chart. 
Each bar represents the probability of asthma diagnosis for individual 
factors and their combinations.

```python
# Convert probabilities dictionary into a DataFrame
probabilities_df = pd.DataFrame.from_dict(probabilities, orient='index', columns=['Probability'])
probabilities_df.reset_index(inplace=True)
probabilities_df.rename(columns={'index': 'Factor'}, inplace=True)

# Convert probabilities to percentages
probabilities_df['Probability'] = probabilities_df['Probability'] * 100

# Sort probabilities in descending order
probabilities_df.sort_values(by='Probability', ascending=False, inplace=True)

# Plot the bar chart
plt.figure(figsize=(12, 8))
bars = plt.bar(probabilities_df['Factor'], probabilities_df['Probability'], color='skyblue', edgecolor='black')

# Add value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, yval + 0.5, f'{yval:.1f}%', ha='center', va='bottom')

# Customize the chart
plt.title('Probability of Asthma Diagnosis for Categorical Factors and Their Combinations')
plt.ylabel('Probability (%)')
plt.xlabel('Factors')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.savefig("Probability Categorical Factors and Their Combinations.png", format="png", dpi=300)
plt.show()
```

#### Summary
We successfully calculated the probabilities of asthma diagnosis for both 
individual categorical factors and their combinations. The bar chart 
highlights the most influential factors, with some combinations showing significantly 
higher probabilities compared to individual factors (eg. Smoking & Gastroesophageal Reflux). This further motivates me to investigate combination effect of more factors.

## Part 2.c Investigating Non-Linear Relationships
This section investigates potential non-linear relationships between continuous factors

#### Scatter Plots with Polynomial Trend Lines
We will visualize scatter plots for each continuous factor against asthma diagnosis 
and fit a polynomial regression line (degree 2) to detect potential non-linear relationships.


```python
import seaborn as sns
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# List of continuous factors
continuous_factors = [
    'PhysicalActivity', 'DietQuality', 'SleepQuality',
    'PollutionExposure', 'PollenExposure', 'DustExposure'
]

# Prepare diagnosis data for non-linear regression
data['Diagnosis_numeric'] = data['Diagnosis'].astype(int)

# Plot scatter plots with polynomial trend lines
plt.figure(figsize=(18, 12))
for i, factor in enumerate(continuous_factors):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=data[factor], y=data['Diagnosis_numeric'], alpha=0.5)
    
    # Fit a polynomial regression model (degree=2 for quadratic relationship)
    poly = PolynomialFeatures(degree=2)
    X_poly = poly.fit_transform(data[factor].values.reshape(-1, 1))
    model = LinearRegression()
    model.fit(X_poly, data['Diagnosis_numeric'])
    
    # Generate prediction curve
    x_range = np.linspace(data[factor].min(), data[factor].max(), 100).reshape(-1, 1)
    x_range_poly = poly.transform(x_range)
    y_pred = model.predict(x_range_poly)
    plt.plot(x_range, y_pred, color='red', label='Polynomial Trend Line')
    
    plt.ylim(0, 0.1)  # Set y-axis limit to 0.2
    plt.title(f'{factor} vs Asthma Diagnosis')
    plt.xlabel(factor)
    plt.ylabel('Asthma Diagnosis')
    plt.legend()

plt.tight_layout()
plt.show()

plt.savefig("Non-linear Relationships for Continuous Factors vs. Asthma Diagnosis.png", format="png", dpi=300)
```

####  Spearman’s Correlation
We will calculate Spearman’s correlation, a rank-based metric, to detect monotonic but non-linear 
relationships between continuous factors and asthma diagnosis.

```python
# Calculate Spearman's correlation for continuous factors with diagnosis
spearman_corr = data[continuous_factors + ['Diagnosis_numeric']].corr(method='spearman')

# Display Spearman's correlation with a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(spearman_corr[['Diagnosis_numeric']].drop('Diagnosis_numeric').sort_values(by='Diagnosis_numeric', ascending=False),
            annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title("Spearman's Correlation with Asthma Diagnosis")
plt.show()

plt.savefig("Spearman's Correlation with Asthma Diagnosis.png", format="png", dpi=300)
```

#### Random Forest Feature Importance
We will train a Random Forest model to evaluate the importance of each continuous factor 
in predicting asthma diagnosis. This provides insights into non-linear interactions.


```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Prepare data for modeling
X = data[continuous_factors]
y = data['Diagnosis_numeric']

# Train a random forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X, y)

# Get feature importances
feature_importances = pd.DataFrame({
    'Factor': continuous_factors,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Plot feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Factor', data=feature_importances, palette='viridis')
plt.title('Feature Importance from Random Forest')
plt.xlabel('Importance Score')
plt.ylabel('Continuous Factors')
plt.tight_layout()
plt.show()

plt.savefig("Feature Importance from Random Forest.png", format="png", dpi=300)
```

## Part 3: Interaction Effects Between Multiple Factors

To explore how interactions between factors affect asthma diagnosis, we will:
1. Balance the dataset using SMOTE to handle class imbalance.
2. Generate interaction terms using `PolynomialFeatures`.
3. Fit a Logistic Regression model to evaluate the significance of these interactions.
4. Visualize the top interaction effects using a heatmap.



#### Balancing the Dataset
We use SMOTE (Synthetic Minority Oversampling Technique) to balance the dataset by oversampling the minority class.


```python
!pip install imbalanced-learn
```

```python
# Import necessary libraries
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Define continuous and binary factors
continuous_factors = [
    'PhysicalActivity', 'DietQuality', 'SleepQuality',
    'PollutionExposure', 'PollenExposure', 'DustExposure'
]
binary_factors = [
    'PetAllergy', 'Smoking', 'FamilyHistoryAsthma',
    'HistoryOfAllergies', 'Eczema', 'HayFever', 'GastroesophagealReflux'
]

# Combine features
selected_features = continuous_factors + binary_factors

# Split data into features (X) and target (y)
X = data[selected_features]  # Features (continuous + binary factors)
y = data['Diagnosis']        # Target

# Balance the dataset
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)

# Verify the balance
print("Class Distribution After SMOTE:")
print(y_balanced.value_counts())
```

#### Adding Interaction Terms
We use `PolynomialFeatures` to generate interaction terms for all pairs of features. Interaction terms capture the combined effect of two factors on asthma diagnosis.


```python
# Import PolynomialFeatures
from sklearn.preprocessing import PolynomialFeatures

# Generate interaction terms
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_poly = poly.fit_transform(X_balanced)

# Get feature names for interaction terms
interaction_features = poly.get_feature_names_out(selected_features)

# Create a DataFrame for the interaction terms
interaction_df = pd.DataFrame(X_poly, columns=interaction_features)

# Display the first few rows
interaction_df.head()
```

#### Fitting Logistic Regression with Interaction Terms
We fit a Logistic Regression model with `class_weight='balanced'` to handle any remaining imbalance and analyze the coefficients for interaction terms.
